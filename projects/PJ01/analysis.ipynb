# Hypothesis: I believe that computer science majors spend more time online for work related reasons than non-computer science majors and that non-computer science majors also spend more time online for social reasons than computer science majors.

## First I will read the `csv` and put the data into a list of dictionaries.

```python
from data_utils import read_csv_rows

data_rows: list[dict[str, str]] = read_csv_rows(DATA_FILE_PATH)
```

## I will separate the data into two groups based on major.

```python
data_cols_comp: list[dict[str, str]] = []
data_cols_other: list[dict[str, str]] = []

for row in data_rows:
    if row["primary_major"] == "Computer Science":
        data_cols_comp.append(row)
    else:
        data_cols_other.append(row)

```

## Next, I will tabulate the data for hours online for Computer Science majors.

```python
from data_utils import head
from tabulate import tabulate
from data_utils import select
from data_utils import columnar

data_cols_comp_column: dict[str, list[str]] = select(columnar(data_cols_comp), ["primary_major", "hours_online_social", "hours_online_work"])

data_cols_comp_head: dict[str, list[str]] = head(data_cols_comp_column, 4)

tabulate(data_cols_comp_head, data_cols_comp_head.keys(), "html")
```

## I will also tabulate the data for hours online for non-Computer Science majors.

```python
from data_utils import head
from tabulate import tabulate
from data_utils import select

data_cols_other_column: dict[str, list[str]] = select(columnar(data_cols_other), ["primary_major", "hours_online_social", "hours_online_work"])

data_cols_other_head: dict[str, list[str]] = head(data_cols_other_column, len(data_cols_other_column))

tabulate(data_cols_other_column, data_cols_other_head.keys(), "html")
```

## Next, I determine the number of students in each range of time spend online and whether it is for social media or work purposes. Then, I will convert that number into a percentage of all students in that same categeory of major and reasons for being online.

```python
from data_utils import count
comp_social_hours: dict[str, int] = count(data_cols_comp_column["hours_online_social"])
comp_work_hours: dict[str, int] = count(data_cols_comp_column["hours_online_work"])

other_social_hours: dict[str, int] = count(data_cols_other_column["hours_online_social"])
other_work_hours: dict[str, int] = count(data_cols_other_column["hours_online_work"])

average_comp_social: dict[str, float] = {}

def averages_social(given_dict: dict[str, int], length_variable: dict[str, list[str]]) -> dict[str, float]:
    new_dict: dict[str, float] = {}

    for value in given_dict:
        new_dict[value] = given_dict[value] / len(length_variable["primary_major"])

    return new_dict

def averages_work(given_dict: dict[str, int], length_variable: dict[str, list[str]]) -> dict[str, float]:
    new_dict: dict[str, float] = {}

    for value in given_dict:
        new_dict[value] = given_dict[value] / len(length_variable["primary_major"])

    return new_dict
    

comp_average_social: dict[str, float] = averages_social(comp_social_hours, data_cols_comp_column)
comp_average_work: dict[str, float] = averages_work(comp_work_hours, data_cols_comp_column)
other_average_social: dict[str, float] = averages_social(other_social_hours, data_cols_other_column)
other_average_work: dict[str, float] = averages_work(other_work_hours, data_cols_other_column)
```

## I will then organize the data in each of the above dictionaries based on hours spent online.

```python
def order_dict(given_dict: dict[str, float]) -> dict[str, str]:
    new_dict: dict[str, str] = {}
    for value in given_dict:
        if "None" in value:
            new_dict[value] = str(given_dict[value])
    
    if "None" not in new_dict:
        new_dict["None"] = "0"


    for value in given_dict:
        if "0 to 2 hours" in value:
            new_dict[value] = str(given_dict[value])
        
    if "0 to 2 hours" not in new_dict:
        new_dict["0 to 2 hours"] = "0"

    for value in given_dict:
        if "3 to 5 hours" in value:
            new_dict[value] = str(given_dict[value])

    if "3 to 5 hours" not in new_dict:
        new_dict["3 to 5 hours"] = "0"

    for value in given_dict:
        if "5 to 10 hours" == value:
            new_dict[value] = str(given_dict[value])

    if "5 to 10 hours" not in new_dict:
        new_dict["5 to 10 hours"] = "0"

    for value in given_dict:
        if "10+ hours" == value:
            new_dict[value] = str(given_dict[value])
        
    if "10+ hours" not in new_dict:
        new_dict["10+ hours"] = "0"

    for value in given_dict:
        if '' == value:
            new_dict["N/a"] = str(given_dict[value])
    
    if 'N/a' not in new_dict:
        new_dict["N/a"] = "0"

    return new_dict





if __name__ == "__main__":
    comp_average_work_ordered: dict[str, str] = order_dict(comp_average_work)
    comp_average_social_ordered: dict[str, str] = order_dict(comp_average_social)
    other_average_work_ordered: dict[str, str] = order_dict(other_average_work)
    other_average_social_ordered: dict[str, str] = order_dict(other_average_social)

    print(comp_average_work_ordered)
    print(comp_average_social_ordered)
    print(other_average_work_ordered)
    print(other_average_social_ordered)
```

## I will then combine all the dictionaries into a single dictionary to later be tabulated and easier to visualize.

```python
final_dict: dict[str, list[str]] = {"Major and Type of Online Activity": [], "None": [], "0 to 2 hours": [], "3 to 5 hours": [], "5 to 10 hours": [], "10+ hours": [], "N/a": []}
final_dict["Major and Type of Online Activity"] = ["Computer Science, Social Hours","Non-Computer Science, Social Hours", "Computer Science, Work Hours",  "Non-Computer Science, Work Hours"]

def make_list_of_values(dict_1: dict[str, str], dict_2: dict[str, str], dict_3: dict[str, str], dict_4: dict[str, str], keyword: str) -> list[str]:
    new_list: list[str] = []
    if keyword in dict_1:
        new_list.append(dict_1[keyword])
     
    if keyword in dict_2:
        new_list.append(dict_2[keyword])

    if keyword in dict_3:
        new_list.append(dict_3[keyword])

    if keyword in dict_4:
        new_list.append(dict_4[keyword])

    return new_list


final_dict["None"] = make_list_of_values(comp_average_social_ordered, other_average_social_ordered, comp_average_work_ordered, other_average_work_ordered, "None")
final_dict["0 to 2 hours"] = make_list_of_values(comp_average_social_ordered, other_average_social_ordered, comp_average_work_ordered, other_average_work_ordered, "0 to 2 hours")
final_dict["0 to 2 hours"] = make_list_of_values(comp_average_social_ordered, other_average_social_ordered, comp_average_work_ordered, other_average_work_ordered, "0 to 2 hours")
final_dict["3 to 5 hours"] = make_list_of_values(comp_average_social_ordered, other_average_social_ordered, comp_average_work_ordered, other_average_work_ordered, "3 to 5 hours")
final_dict["5 to 10 hours"] = make_list_of_values(comp_average_social_ordered, other_average_social_ordered, comp_average_work_ordered, other_average_work_ordered, "5 to 10 hours")
final_dict["10+ hours"] = make_list_of_values(comp_average_social_ordered, other_average_social_ordered, comp_average_work_ordered, other_average_work_ordered, "10+ hours")
final_dict["N/a"] = make_list_of_values(comp_average_social_ordered, other_average_social_ordered, comp_average_work_ordered, other_average_work_ordered, "N/a")
```

## Finally, I tabulate the data so I can compare all the percentages spent online for major and activity side-by-side.

```python
from tabulate import tabulate

tabulate(final_dict, final_dict.keys(), "html")
```
